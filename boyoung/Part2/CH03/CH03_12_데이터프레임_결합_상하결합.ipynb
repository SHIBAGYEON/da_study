- 데이터 프레임 결합


1) 상하 결합

```python
import pandas as pd

df1 = pd.DataFrame({'A':[1,2,3], 'B':[11,12,13], 'C':[21,22,23]})
df2 = pd.DataFrame({'A': [4,5,6], 'B':[14,15,16], 'C':[24,25,26]})
```

```python
# 2개 데이터 프레임 결합
# 상하관계 결합 .concat([위, 아래])
pd.concat([df1, df2])
```

```python
# 인덱스는 그대로임
pd.concat([df2, df1])
```

```python
# ignore_index = True 인덱스 초기화
pd.concat([df1, df2], ignore_index = True)
```

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[11,12,13], 'C':[21,22,23]})
df2 = pd.DataFrame({'B':[14,15,16], 'A': [4,5,6], 'C':[24,25,26]})
```

```python
# 컬럼명 확인하여 결합함
pd.concat([df1, df2])
```

- 서로 다른 필드로 구성되어 있는 데이터 프레임 결합

```python
df1 = pd.DataFrame({'A': [1,2,3], 'B':[11,12,13], 'C': [21,22,23], 'D':[31,32,33]})
df2 = pd.DataFrame({'A' : [3,4,5], 'B':[13,14,15], 'C':[23,24,25], 'E':[41,42,43]})
```

```python
df1
```

```python
df2
```

```python

pd.concat([df1, df2])
```

```python
# 합집합 출력 옵션
pd.concat([df1, df2], join = 'outer')
```

```python
# 교집합 출력 옵션
pd.concat([df1, df2], join = 'inner')
```
