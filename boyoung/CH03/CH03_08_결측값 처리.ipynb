**결측값 처리**

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a':[1, 1, 3, 4, 5], 'b':[2, 3, np.nan, 3, 4], 'c':[3, 4, 7, 6, 4]})
```

```python
df
```

- 결측 유무 확인

```python
# 해당 열에 결측 값이 있는지 확인 가능
df.isnull()
```

- 결측값 개수 확인

```python
# 해당 열에 결측값이 몇개 있는지?
# True = 1, False = 0 의 합을 구해줌
df.isnull().sum()
```

- 결측값이 포함된 열/행 지우기

```python
# 결측값이 포함된 행 지우기
df.dropna()
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
# 결측값이 포함된 열 지우기
```

```python
df = pd.DataFrame({'a':[1, 1, 3, 4, 5], 'b':[2, 3, np.nan, 3, 4], 'c':[3, 4, 7, 6, 4]})
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
df.dropna(axis=1, inplace = True)
```

```python
df
```

- 결측값을 다른 값으로 대체하기

```python
df = pd.DataFrame({'a':[1,1,3,4,5], 'b':[2,3,np.nan,3,4], 'c':[3,4,7,6,4]})
```

```python
df
```

- 특정값으로 대체하기

```python
# 결측값을 어떤값으로 바꿀껀지 ()안에 넣으면 됨
df.fillna(0, inplace = True)
```

```python
df
```

- 앞이나 뒤의 숫자로 바꿔주기

```python
df = pd.DataFrame({'a':[1,1,3,4,np.nan], 'b':[2,3,np.nan,np.nan,4], 'c':[np.nan,4,1,1,4]})
```

```python
df
```

1) 뒤의 값으로 채우기

```python
df.fillna(method='bfill')
```

```python
df
```

2) 앞의 값으로 채우기

```python
# c열 맨앞값이 결측치여서 대체해서 가져올 값이 없음
df.fillna(method='ffill')
```

- limit 설정

```python
df
```

```python
# limit = 1 앞에있는 값을 1번만 가져오겠다
df.fillna(method='ffill', limit=1)
```

- 문제: 데이터 프레임에 존재하는 결측값들을 뒤의 값으로 대체한 이후 앞의 값으로 대체하시오.

```python
df = pd.DataFrame({'a':[1,1,3,4,np.nan], 'b':[2,3,np.nan,np.nan,4], 'c':[np.nan,4,1,1,4]})
```

```python
df
```

```python
df.fillna(method = 'bfill', inplace = True)
```

```python
df
```

```python
df.fillna(method = 'ffill', inplace = True)
```

```python
df
```

- 평균으로 대체

```python
df = pd.DataFrame({'a':[1,1,3,4,np.nan], 'b':[2,3,np.nan,np.nan,4], 'c':[np.nan,4,1,1,4]})
```

```python
df
```

```python
# 특정열 평균값 구하기
df.mean()['a']
```

```python
df.fillna(df.mean()['a'], inplace = True)
```

```python
df
```

```python
# 전체 평균 구하기
df.mean()
```

```python
# 각 열의 평균을 결측값에 대입해줌!!
df.fillna(df.mean())
```

```python
df
```

- 문제: b와 c의 결측값을 데이터 프레임의 전체 값의 평균값으로 치환하시오.

```python
df.fillna(df.mean()[['a','b']])
```
