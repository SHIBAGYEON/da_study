**loc 과 iloc을 이용한 원하는 위치의 데이터 추출**

- 데이터 생성

```python
import pandas as pd
df = pd.DataFrame({'a':[i for i in range(1,11)],'b':[i for i in range(11,21)], 'c': [i for i in range(21,31)]})
```

```python
df
```

- 문제: a,b 열을 추출하시오.

```python
df['a','b']
```

```python
# series 생성시 [[]] 대괄호 2개 사용하면 2개 컬럼 추출 가능
df[['a','b']]
```

```python
type(df[['a','b']])
```

- 첫 번째 행의 데이터를 추출하시오

```python
# 컬럼을 추출하는 데이터이므로, 행 추출 불가
df[0]
```

- loc을 통한 해결

```python
df.loc[0]
```

- 슬라이스도 가능



---


loc와 리스트의 차이점

1. 일반적인 리스트 슬라이스 할때는 끝자리수는 추출 X (-1까지 자리수만 추출함)
2. loc함수는 해당 인덱스 값을 다 가져옴!!!



---





```python
df.loc[2:4]
```

- 인덱스가 문자로 이루어진 데이터 프레임 생성

```python
index = ['a','b','c','d','e','f','g','g','h','i']
```

```python
df = pd.DataFrame({'a':[i for i in range(1,11)], 'b': [i for i in range(11,21)], 'c':[i for i in range(21,31)]}, index = index)
```

```python
df
```

```python
# 0이라는 인덱스 존재하지 X 에러나옴
df.loc[0]
```

```python
# 인덱스 g 2개가 주어졌을때 loc는 모두 가져옴
df.loc['g']
```

```python
# 저장된 순서대로 출력하는지? 알파벳 순서대로 출력하는지?
# 데이터가 쌓여진 순서대로 출력함!!!
df.loc['c':] 
```

- 문제: 열이 a,c 이며 인덱스가 g,i 인 데이터를 출력하시오.

```python
# 행의 위치, 열의 위치 넣으면됨
df.loc[['g','i'],['a','c']]
```

- 문제: 처음부터 5번째 까지의 데이터와 첫 번째 열과 세번째 열의 데이터를 추출하시오.

```python
df = pd.DataFrame({'a':[i for i in range(1,11)], 'b':[i for i in range(11,21)], 'c':[i for i in range(21,31)]})
```

```python
df
```

- iloc을 이용한 해결

```python
# 행, 열 순서로 기재
df.iloc[:5,[0,2]]
```

- 문제: 처음부터 5번째 까지의 데이터와 첫 번째 열과 세번째 열의 데이터를 추출하시오.

```python
index = ['a','b','c','d','e','f','g','g','h','i']
df = pd.DataFrame({'a': [i for i in range(1,11)], 'b': [i for i in range(11,21)], 'c':[i for i in range(21,31)]}, index = index)
```

```python
df
```

```python
df.iloc[:5,[0,2]]
```
