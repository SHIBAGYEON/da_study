# boxplot을 이용한 상자 수염 그림 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10,5)})
import pandas as pd
```

```python
# iris data set 로드
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# boxplot 그리기
sns.boxplot(data = df)
# 점들은 이상치, 아래 줄은 최소값, 위 줄은 최대값
```

```python
# 새로방향으로 그림 크기 키우기
plt.subplots(figsize = (7,8))
sns.boxplot(data = df)
```

```python
# 가로방향으로 그리기
```

```python
sns.boxplot(data = df, orient = 'h')
```

- swarmplot을 이용한 raw data 확인

```python
plt.subplots(figsize = (10, 10))
sns.swarmplot(data = df)
```

- boxplot과 swarmplot 겹쳐그리기

```python
plt.subplots(figsize = (10, 10))
sns.boxplot(data = df)
sns.swarmplot(data = df)
```

```python
plt.subplots(figsize = (10, 10))
sns.boxplot(data = df)
sns.swarmplot(data = df, color = 'black')
```

- violinplot

```python
sns.violinplot(data = df)
```

```python
sns.stripplot(data = df)
```
