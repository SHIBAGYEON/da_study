# heatmap을 이용한 히트맵 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(10,5)})
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

```python
df.describe()
```

```python
# 결측값 확인
df.isnull().sum()
```

- 문제: 각 년도별 월별 승객수를 히트맵으로 나타내시오.

```python
# pivot을 이용한 데이터 형태 변경
pivot_data = df.pivot("month", "year", "passengers")
```

```python
pivot_data
```

```python
sns.heatmap(pivot_data)
```

```python
# 구분선 추가
sns.heatmap(pivot_data, linewidths = 5)
```

```python
sns.heatmap(pivot_data, cmap = 'Blues')
```

```python
# 수치 입력
sns.heatmap(pivot_data, cmap='Blues', annot = True)
```

```python
#정수 형태로 출력
sns.heatmap(pivot_data, cmap='Blues', annot = True, fmt = 'd')
```

```python
# colorbar 가로로 놓기
fig, (ax, cbar_ax) = plt.subplots(2) # 
```

```python
ax
```

```python
cbar_ax
```

```python
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation":"horizontal"}) # 방향은 수평
```

```python
fig
```

```python
# 수직으로 변경
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation":"vertical"}) # 방향은 수평
```

```python
fig
```

```python
# color bar 와 히트맵 값의 비율 조정
grid_kws = {"height_rations":(.85,.1), "hsapce": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation":"horizontal"}) # 방향은 수평
```

```python
# 최종
grid_kws = {"height_rations":(.85,.1), "hsapce": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation":"horizontal"}) # 방향은 수평
```
