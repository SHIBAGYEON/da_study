# 여러 데이터를 한 차트에 그리기

```python
import matplotlib.pyplot as plt
```

- 문제: data1(x:1,2,3, y:1,2,3)과 data2(x:1,2,3, y:1,4,7)을 그래프로 출력하시오.

```python
x1= [1,2,3]
y1= [1,2,3]
x2= [1,2,3]
y2= [1,4,7]
```

```python
plt.plot(x1, y1, color = 'blue') 
plt.plot(x2, y2, color = 'red')   
```

또는

```python
# 한줄로 표현
plt.plot(x1,y1,'b',x2,y2,'r')
```

- 범례추가

```python
x1= [1,2,3]
y1= [1,2,3]
x2= [1,2,3]
y2= [1,4,7]
```

```python
plt.plot(x1, y1, color='b', label='data1')
plt.plot(x2, y2, color='r', label='data2')
plt.legend()
```

또는

```python
plt.plot(x1, y1, color = 'blue')
plt.plot(x2, y2, color = 'red')
plt.legend(['data1', 'data2'])
```

- 범례 위치 / 크기 설정

```python
plt.plot(x1, y1, color = 'blue')
plt.plot(x2, y2, color = 'red')
plt.legend(['data1', 'data2'], loc = 'upper right')
```

- 범례 폰트 사이즈 변경

```python
plt.plot(x1, y1, color = 'blue')
plt.plot(x2, y2, color = 'red')
plt.legend(['data1', 'data2'], fontsize = 20)
```

- 문제: data1(x:1,2,3, y:1,2,3)과 data2(x:1,2,3, y:1,100,200)을 그래프로 출력하시오.

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

```python
plt.plot(x1, y1, color='blue') # 파란색줄 데이터 표시가 잘 안됨
plt.plot(x2, y2, color= 'red')
```
