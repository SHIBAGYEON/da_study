**정렬**

- 데이터 생성

```python
import pandas as pd

df = pd.DataFrame({'a':[2,3,2,7,4], 'b':[2,1,3,5,3], 'c':[1,1,2,3,5]})
```

```python
df
```

- 인덱스 기준 정렬

```python
df.sort_index()
```

- 내림차순일때는 ascending = False

```python
df.sort_index(ascending = False)
```

```python
df
```

- 결과를 저장하고 싶으면 inplace = True

```python
df.sort_index(ascending =  False, inplace = True)
```

```python
df
```

- 인덱스 초기화는 reset_index()

```python
df.reset_index()
```

```python
# reset인덱스 저장이 안됨!!
df
```

```python
# reset index 결과값에서 index 없애기 
df. reset_index(drop = True)
```

```python
df. reset_index(drop = True, inplace = True)
```

```python
df
```

- 값 기준 정렬

```python
df =  pd.DataFrame({'a':[2,3,2,7,4], 'b':[2,1,3,5,3], 'c':[1,1,2,3,5]})
```

```python
df
```

- a열 기준으로 오름차순 정렬하시오.

```python
df.sort_values(by = ['a'])
```

```python
df
```

```python
# inplace = True 이용하여 고정시켜야함
df.sort_values(by = ['a'], inplace = True)
```

```python
df
```

- a 열 기준으로 내림차순 정렬하시오.

```python
df =  pd.DataFrame({'a':[2,3,2,7,4], 'b':[2,1,3,5,3], 'c':[1,1,2,3,5]})
```

```python
df.sort_values(by = ['a'], ascending = False)
```

- 문제: a,b 열 기준으로 오름차순 정렬하시오.

```python
df.sort_values(by = ['a','b'])
```

```python
df.sort_values(by = ['a','b'], ascending = [True, False])
```

```python
df
```

```python
df.sort_values(by = ['a','b'], ascending = [True, False], inplace = True)
```

```python
df
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df
```
