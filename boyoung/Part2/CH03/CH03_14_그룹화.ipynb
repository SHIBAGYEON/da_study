# - 그룹화

```python
import pandas as pd
df1 = pd.DataFrame({'ID':[1,2,3,4,5], '가입일':['2021-01-02', '2021-01-04', '2021-01-10', '2021-02-10', '2021-02-24'], '성별':['F','M','F','M','M']})
df2 = pd.DataFrame({'구매순서':[1,2,3,4,5], 'ID':[1,1,2,4,1], '구매월':[1,1,2,2,3], '금액':[1000,1500,2000,3000,4000]})
```

- 문제: df1 은 회원의 정보를 저장한 데이터 프레임이며, df2는 각 회원의 구매내역을 저장한 데이터 프레임이다. 각 회원의 정보와 구매내역을 취합하여 하나의 데이터 프레임을 만드시오.

```python
pd.merge(df1,df2, how='left', on='ID')
```

- 문제: df1 은 회원의 정보를 저장한 데이터 프레임이며, df2는 각 회원의 구매내역을 저장한 데이터 프레임이다. 각 회원의 누적 금액을 회원 ID별로 구하시오.

```python
df2
```

```python
# df2 기준으로 그룹화를 할건데, 기준은 ID이고, 그 결과로 각 ID별 합계 추출
df2.groupby(by =['ID'])['금액'].sum()
```

```python
type(df2.groupby(by =['ID'])['금액'].sum())
```

```python
s2 = df2.groupby(by =['ID'])['금액'].sum()
```

```python
pd.merge(df1, s2, how = 'left', on = 'ID')
```

```python
df2.groupby(['ID','구매월'], as_index = False)['금액'].sum()
```

```python
type(df2.groupby(['ID','구매월'], as_index = False)['금액'].sum())
```

```python
pd.merge(df1, df3, how= 'left', on = 'ID')
```

- 문제: df는 각 회원의 구매내역을 저장한 데이터 프레임이다. 각 회원의 누적 금액과 누적 구매횟수를 회원 ID별로 구하시오.

```python
df = pd.DataFrame({'구매순서':[1,2,3,4,5], 'ID':[1,1,2,4,1], '구매월':[1,1,2,2,3], '금액': [1000,1500,2000,3000, 4000], '수수료':[100,150,200,300,400]})
```

```python
df
```

```python
# agg 여러가지 결과를 함께 확인 가능
df.groupby(['ID'])['금액'].agg([sum, len])
```

```python
# agg와 as_index = False는 사용불가
df.groupby(['ID'], as_index = False)['금액'].agg([sum, len])
```

```python
# 인덱스 초기화 함수 .reset_index()
df2.reset_index(inplace = True)
```

```python
df2
```

- 문제: df는 각 회원의 구매 내역을 저장한 데이터 프레임이다. 각 회원의 최대 사용 금액 / 최소 사용금액과 최저 수수료의 값을 구하시오.

```python
df
```

```python
# ID기준으로 그룹화하고, 금액 최고, 최소값, 수수료는 최저값 구하기
df.groupby(['ID']).agg({'금액':[max,min], '수수료':[min]})
```

```python
df2 = df.groupby(['ID']).agg({'금액':[max,min], '수수료':[min]})
```

```python
df2.reset_index()
```

```python
# 칼럼명 뽑기
df2.columns
```

```python
# 값출력
df2.columns.values
```

```python
# join 함수=> 각 내부에 포함된 값을 하나로 합쳐줌(어렵다....)
df2.columns = ['_'.join(col) for col in df2.columns.values]
```
