# **머신러닝 이란?**

**1. 머신러닝 이란?**
- 머신러닝 : 알고리즘을 개발하는 연구 분야, 기계가 일일이 코드로 명시하지 않은 동작을 데이터로부터 학습하여 실행할 수 있도록 함
> 데이터 / 학습/ 알고리즘
> - 데이터: 객관적인 기준을 위한 정답과 요인들
> - 학습: 데이터로부터 최적의 값을 찾는 과정
> - 알고리즘: 문제를 풀기위한 의사결정 과정

**2. 머신러닝 알고리즘 종류**

2-1. 지도학습: 정답이 있는 문제를 풀 경우
- 회귀(연속성 변수): 값이 정수처럼 명확하지 않음 ex. 키, 몸무게…
프렌시스 골턴 부모와 자녀의 키사이에는 선형적인 관계가 있고 키가 커지거나 작아지는 것 보다는 전체 키 평균으로 돌아가려는 경향이 있다는 가설
Ex. 어떤 그래프에 점들이 있고, 점들 예측하는 선을 그음 > 데이터 예측값 확인 가능

- 분류(비연속형 변수): 변주형 변수, 값이 정수처럼 정확함 ex. 성별, 도시, iris꽃 종류


2-2. 비지도학습: 정답이 없는 문제를 풀 경우
- 군집분석: 주어진 데이터가 어떻게 구성 되었는지 알아내려는 분석 방법
- 강화학습: 행동에 따른 보상을 최대화 시키는 분석방법

**3. Model Selection**

3-1. 모델
x가 주어졌을때 f라는 함수를 통해 y라는 값을 도출하는 과정
이때 함수 f를 모델 또는 알고리즘이라고 부른다.

> <모델의 수식>
> * Y = F(X)
> * X: 데이터, y: 예측값

- 모델의 목적: 데이터를 이용해 값을 예측
- 모델의 평가: 모델이 값을 잘 예측하는지 평가

3-2.  데이터
- Train data: 학습에 사용되는 데이터
- Test data: 학습에 사용되지 않은 데이터, 모델이 실제로 잘 예측하는지 알기 위해서는 학습에 사용되지 않은 데이터를 이용해 평가해야함

3-3. 과대적합, 과소적합
- Underfitting(과소 적합): fitting이 잘 되지 않음. 학습데이터를 잘 맞추지 못하는 현상
확인하는 방법: 
>Train data로 학습된 모델을 Train data로 평가한다\
Train data를 잘 맞추지 못한다면 Underfitting 상태임

- Overfitting(과대 적합): 학습데이터는 잘 맞추지만 학습데이터 외에는 잘 맞추지못함
확인하는 방법:
>Train data을 잘 학습한 모델을 Test data로 평가한다\
Train data는 잘 맞추지만 Test data를 잘 맞추지 못한다면 Overfitting 상태임


Undorfittod.png

- Data split: 데이터를 Train data와 Test data로 나누는 것\
Train Data : Test Data 7:3\
Train Data : Valid Data : Test Data 
Valid data? 학습이 완료된 모델을 검증하기 위한 데이터, 학습에 직접적으로 사용되지  않지만 학습에 관여하는 데이터임

- Test data: 최종 모델 성능 검증하기 위한 데이터
학습에도 사용하지 않으면서 관여하지 않는 데이터임
> Valid와 Test 차이점: 학습에 관여하는지 여부




3-4. Cross Validation\
Valid data overfitting 문제 발생 가능 >> Valid 데이터 고정하지 않고, 계속 변경함으로써 Overfitting되는 것을 막기 위한 방법

- Cross Validation 종류
1. LOOCV(Leave One Out Cross Validation)\
하나의 데이터를 제외하고 모델을 학습한후 평가\
단점: 데이터 개수만큼 모델 학습 해야되므로, 데이터가 많을경우 시간이 오래걸림


LOOCV (Leave One Out Cross Validation).png

2. K-fold\
데이터를 K개로 분할한 후 한개의 분할 데이터를 제외한 후 학습에 사용\
제외된 데이터는 학습이 완료된 후 평가에 사용
> LOOCV n개, k-fold는 k개의 평가지표가 생성됨\
생성된 평가지표의 평균을 이용해 모델의 성능을 평가\
마지막 가장 좋은 성능 모델로 전체 Train 데이터를 이용해 모델 학습
