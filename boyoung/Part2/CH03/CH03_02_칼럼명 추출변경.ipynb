- rename을 통한 칼럼명 변경

```python
# 데이터 프레임 재생성
df = pd.DataFrame({'a':[1,2,3],'b':[4,5,6],'c':[7,8,9]})
# 치환 변경
df.columns = ['d','e','f']
```

```python
# df.rename(columns = {기존 칼럼명:변경 칼럼명})
df.rename(columns = {'d':'디', 'f':'에프'})
```

```python
df
```

```python
# inplace = True 로 되어있어야 컬럼 변경내용 저장됨
df.rename(columns = {'d':'디', 'f':'에프'}, inplace = True)
```

```python
df
```
