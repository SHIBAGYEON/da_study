```python
import pandas as pd
df = pd.DataFrame({'a':[1,2,3,4,5]})
```

- 문제: a가 2보다 작으면 '2'미만, 4보다 작으면 '4미만', 4보다 크면 '4이상'이 저장된 b 칼럼을 추가하시오.

```python
df
```

```python
df['b'] = 0
```

```python
df
```

```python
# a라는 변수에 첫번째 조건을 만족하는 데이터 프레임
a = df[df['a'] < 2]
```

```python
a
```

```python
df['b'][a.index]='2미만'
```

```python
df
```

```python
a = df[(df['a'] >= 2)& (df['a'] < 4)]
```

```python
a
```

```python
# 데이터프레임 일부 복사하고 값을 수정하면 발생하는 에러
df['b'][a.index] = '4미만'
```

```python
# 해당 코드입력하여 경고문 없앨수 있음
pd.set_option('mode.chained_assignment', None)
```

```python
df
```

```python
a = df[df['a'] >= 4]
```

```python
a
```

```python
# df의 b중에 0번째 값을 4이상이라고 바꿔 주어라.
df['b'][a.index] = '4이상'
```

- 함수 + apply를 이용한 해결

```python
# 사용자 정의 함수
def case_function(x):
  if x < 2:
    return '2 미만'
  elif x < 4:
    return '4 미만'
  else:
    return '4 이상'
```

```python
# df['a']의 위에서 적용한 함수를 적용하여라
df['c'] = df['a'].apply(case_function)
```

```python
df
```

- 문제: a가 1이면 'one', 2이면 'two', 3이면 'three', 4이면 'four', 5이면 'five'를 출력하는 칼럼d를 만드시오.

```python
df
```

```python
def case_function2(x):
  if x == 1:
    return 'one'
  elif x == 2:
    return 'two'
  elif x == 3:
    return 'three'
  elif x == 4:
    return 'four'
  else: # else 뒤에 조건 X 
    return 'five'
```

```python
df['d'] = df['a'].apply(case_function2)
```

```python
df
```
