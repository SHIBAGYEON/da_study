copy를 이용한 데이터복사

- 데이터 생성

```python
import pandas as pd
df = pd.DataFrame({'a':[1,2,3],'b':[4,5,6],'c':[7,8,9]})
```

- 문제: 필드명을 a,b,c에서 d,e,f로 변경하시오

```python
df.columns=['d','e','f']
```

```python
df
```

- 문제: 필드명 a를 '에이'로 변경하시오.

```python
df = pd.DataFrame({'a':[1,2,3],'b':[4,5,6],'c':[7,8,9]})
```

```python
df.rename(columns = {'a':'에이'}, inplace = True)
```

```python
df
```

- deep copy를 통한 해결

```python
import copy
```

```python
df = pd.DataFrame({'a':[1,2,3],'b':[4,5,6],'c':[7,8,9]})
```

```python
# 각 데이터프레임을 따로 저장
df2 = copy.deepcopy(df)
```

```python
df
```

```python
df2
```

```python
df.columns = ['d','e','f']
```

```python
df
```

```python
df2
```
