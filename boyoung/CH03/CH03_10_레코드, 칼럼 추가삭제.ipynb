**레코드, 칼럼 추가 / 삭제**

```python
import pandas as pd
df = pd.DataFrame({'a': [1, 2, 3, 4, 5], 'b':[2, 3, 2, 3, 4], 'c':[3, 4, 7, 6, 4]})
```

```python
df
```

- 칼럼 추가

- 문제: 1,3,6,4,8 로 이루어진 d 칼럼을 추가하시오.

```python
df['d'] = [1,3,6,4,8]
```

```python
df
```

- 문제: 1로 이루어진 e칼럼을 추가하시오.

```python
df['e'] = [1, 1, 1, 1, 1]
```

또는

```python
df['e'] = 1
```

```python
df
```

```python
df.dtypes
```

- 문제: a + b - c 의 결과로 이루어진 f 칼럼을 추가하시오.



```python
df['f'] = df['a'] + df['b'] - df['c']
```

```python
df
```

- 칼럼 삭제


> 문제: 칼럼 d,e,f를 삭제하시오.



```python
# axis = 1 >> 열 칼럼 기준으로 삭제
df.drop(['d','e','f'], axis =1)
```

```python
df
```

```python
df.drop(['d','e','f'], axis =1, inplace = True)
```

```python
df
```

- 레코드 추가


> - 문제: a에는 6, b에는 7, c에는 8을 추가하시오.



```python
# ignore_index=True 옵션을 추가해야됨
df.append({'a':6, 'b':7, 'c':8})
```

```python
df.append({'a':6, 'b':7, 'c':8}, ignore_index = True)
```

```python
df
```

```python
# 해당 df 변수명에 덮어쓰기 실행
df = df.append({'a':6, 'b':7, 'c':8}, ignore_index = True)
df
```

- a 에는 7, b에는 8, c에는 9를 추가하시오.

```python
# 새로운 index 6이라는 값에 7,8,9를 추가
df.loc[6] = [7,8,9]
```

```python
df
```

- 레코드 삭제


> 문제: 첫 번째 레코드를 삭제하시오.



```python
df
```

```python
# df에서 0 행을 삭제하여라
df.drop(0)
```

```python
# 변수명 덮어써서 저장해야 변경사항이 반영됨
df
```

```python
# .drop(0~1까지 삭제)
df = df.drop([0,1])
```

```python
df
```

- 문제: 첫 번째에서 네번째 레코드를 삭제하시오.

```python
df = pd.DataFrame({'a': [1,1,3,4,5], 'b':[2,3,2,3,4], 'c':[3,4,7,6,4]})
```

```python
df
```

```python
# 리스트 컴프리핸션 이용
df.drop([i for i in range(4)])
```

```python
df
```

또는

```python
# .index[:4] 0~4까지의 값을 뽑아내서 삭제
df.drop(df.index[:4])
```

- 문제: a가 4미만인 레코드들을 삭제하시오.

```python
df
```

```python
# df의 a가 4 미만인 것을 가져옴
df[df['a']>4].index
```

```python
df.drop(df[df['a']<4].index)
```

- 문제: a가 3 미만이고 c가 4인 레코드들을 삭제하시오. 

```python
df
```

```python
df[(df['a'] < 3) & (df['c'] == 4)].index
```

```python
df.drop(df[(df['a'] < 3) & (df['c'] == 4)].index)
```
