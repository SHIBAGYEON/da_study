3. 최적화 알고리즘

3-1. Bisection Method
- 임의의 두개의 값을 설정한다
- 두 값의 y값을 비교한다
- y 값이 큰 점을 두점의 가운데 점으로 바꾼다
- 임의의 두 값의 차이가 작아질 때 까지 1~3을 반복한다

3-2. Gradient Descent(경사 하강법)\
함수의 기울기(경사)를 구하고 경사의 절대값이 낮은 쪽으로 게속 이동시켜 극값에 이를 때까지 반복시키는 것

Graident Descent 방법
1. 임의의 값 하나를 설정한다
2. 임의의 값에서 기울기를 계산한다
3. 기울기와 Learning rate를 곱한 값을 임의의 값에서 뺀다
4. 기울기가 0에 가까워질 때 까지 1~3번을 반복한다.
