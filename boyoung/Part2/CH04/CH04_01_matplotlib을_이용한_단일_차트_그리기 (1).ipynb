# matplotlib을 이용한 단일 차트 그리기

- library 호출

```python
import matplotlib.pyplot as plt
```

- 문제: a가 1,2,3이고 y가 4,5,6인 그래프를 그리시오.

```python
x = [1,2,3]
y = [4,5,6]
```

```python
# 그래프 그려줄 도화지 생성
plt.figure() # 생략가능
# 그래프 그리기
plt.plot(x,y)
```

또는

```python
plt.plot(x,y)
```

- lineplot 은 y값만 있으면 그림을 그릴 수 있음
- 이 때, x의 값은 0,1,2 순서로 자동 지정됨

```python
plt.plot(y) # y만 주어졌을때 자동적으로 0부터 값이 커짐
```

- linewidth를 통한 선 굵기 조정

```python
plt.plot(x,y, linewidth = 10)
```

- color를 통한 선 색상 조정

```python
plt.plot(x,y, color='r')
```

- marker을 이용한 데이터 위치 표시

```python
plt.plot(x,y, marker ='o')
```

```python
plt.plot(x,y, marker = 'o', markersize = 10)
```

- 선 형태 변경

```python
plt.plot(x,y, linestyle = ':')
```

- 선 형태 변경

```python
plt.plot(x,y,'o')
```

```python
# 빨간 동그라미 점 표시
plt.plot(x,y, 'ro')
```

- 그래프명

```python
plt.plot(x,y)
plt.title('title')
```

```python
plt.plot(x,y)
plt.title('title', fontsize = 20)
```

- x축 명, y축 명

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

- 축 명 폰트 크기 변경

```python
plt.plot(x, y)
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize = 20)
```

- 그리드 설정(전체)

```python
plt.plot(x, y)
plt.grid(True) # 격자 
```

- 그리드 설정(x축만)

```python
plt.plot(x,y)
plt.grid(True, axis = 'x')
```

- 그리드 설정(y축만)

```python
plt.plot(x,y)
plt.grid(True, axis = 'y')
```

```python
# 특정 구간만 확대해서 보여줌
plt.plot(x,y)
plt.xlim([1,2])
plt.ylim([4,5])
```

- axis을 사용한 범위 설정

```python
plt.plot(x,y)
# plt.axis([x값 최소값/최대값, y값 최소값/최대값])
plt.axis([1,2,4,5])
```

```python
# 눈금 글꼴 크기 변경
plt.plot(x,y)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 17)
```

- 그래프에 텍스트 삽입

```python
plt.plot(x,y)
# 어느점 위치에 텍스트 삽입할 건지
plt.text(2,5, 'test', fontsize ='20')
```
