# **Linear Regression**

- 선형 회귀(Linear Regression)의 정의\
예측값을 직선으로 표현하는 모델

- 데이터\
y : 예측값\
x : 입력값

- 계수\
B0 : 편향\
B1 : 계수

- B0 : 편향\
그래프에서 회귀선의 위치\
x 가 0일때의 예측 값

스크린샷 2022-11-20 오후 7.01.26.png

- B1 : 계수(Coefficient)\
회귀선의 기울기\
x가 1 증가할 때 예측 값이 증가하는 정도

스크린샷 2022-11-20 오후 7.03.50.png

>**어떤 직선이 더 잘 예측할까?**\
실제 값과 예측값의 차이가 작은 직석이 더 잘 예측!

- Mean Square Error\
차이의 크기를 반영하기 위해서 제곱을 하고 평균을 구한다
