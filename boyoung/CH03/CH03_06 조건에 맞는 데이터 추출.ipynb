**조건에 맞는 데이터 추출**

- 데이터 생성

```python
import pandas as pd
df = pd.DataFrame({'a':[i for i in range(1,11)], 'b':[i for i in range(11,21)],'c':[i for i in range(21,31)]})
```

```python
df
```

- 문제: a와 c열을 출력하시오.

```python
df[['a','c']]
```

- 문제: a가 3 이상인 데이터를 출력하시오.

```python
# 데이터프레임[df의 a가 3보다 큰것]
df[df['a']>=3]
```

- 문제: a가 3 이상인 데이터 중 a,c열만 출력하시오.

```python
df[df['a']>=3][['a','c']]
```

- 문제: a가 3 이상이고, b가 16 미만인 데이터를 출력하시오.

```python
df
```

```python
df[(df['a']>=3) & (df['b']<16)]
```

```python
# 조건으로 시리즈 만들기
a = (df['a']>=3) & (df['b']<16)
```

```python
a
```

```python
df[a]
```

- 문제: a가 3이하 이거나 7이상인 데이터를 추출하시오.

```python
df
```

```python
df[(df['a']<=3) | (df['a']>=7)]
```

- 문제: a가 3 이상이고, b가 16미만이거나 c가 30인 데이터를 출력하시오.

```python
df
```

```python
# 큰 조건이 30 이상임, 작은 조건은 괄호 2번침
df[(df['a']>=3) & ((df['b']<16) | (df['c'] == 30))]
```
